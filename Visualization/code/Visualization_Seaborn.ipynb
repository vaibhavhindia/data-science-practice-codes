# Categorical Plot

<li> Focus on distributions using categorical data in reference to one of the numerical columns
<li> Aggregate categorical data based on a function (mean is the default)
<li> Estimate total bill amount based on sex
<li> With estimator you can define functions to use other than the mean like those

### a. bar plot

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
df_used_cars
```

```python
pal = sns.color_palette("viridis")
sns.barplot(x="Brand",y="Cars Listings",data=df_used_cars, palette=pal)
plt.xticks(rotation = 90, fontsize = 10) 
plt.show()
```

### b.Horizontal Barplot

```python
pal = sns.color_palette("pastel")
#sns.set_color_codes("pastel")
sns.barplot(y=df_used_cars["Brand"], x=df_used_cars["Cars Listings"])
plt.xticks(rotation = 90, fontsize = 10) 
plt.show()
```

```python
pal = sns.color_palette("Set2")
#sns.set_palette('winter')
sns.catplot(x="Brand", 
            y='Cars Listings',
            data=df_used_cars,
            kind="bar",
            palette=pal
           )
plt.xticks(rotation=90)
# Show plot
plt.show()
```

```python
# Create count plot with region on the y-axis
sns.countplot(x="day", data=tips, 
              palette=pal,
             hue='sex'
             )

# Show plot
plt.show()
```

```python
plt.figure(figsize = (1, 1)) 

sns.catplot(y="day", 
            data=tips,
            kind="count",
            hue="smoker",
            col='sex'
           )

# Show plot
plt.show()
```

#### b. Boxplot

<li> A box plot allows you to compare different variables
<li> The box shows the quartiles of the data. The bar in the middle is the median and
<li> the box extends 1 standard deviation from the median
<li> The whiskers extend to all the other data aside from the points that are considered to be outliers

```python
sns.catplot(x="day", 
            y="total_bill",
            data=tips, 
            kind="box",
            hue='sex',
            palette=pal
           )

# Show plot
plt.show()
```

### Strip Plot

```python
sns.catplot(x ='day', y ='total_bill', data = tips,
               hue ='sex',  
              kind='strip',
              palette = pal
           )
plt.show()
```

### Swarm Plot

```python
sns.catplot(x ='day', y ='total_bill', data = tips,
               hue ='smoker',  
            kind='swarm',
            palette=pal
           )
plt.show()
```

### Violin Plot

```python
#my_pal = {"Thur": "g", "Fri": "b", "Sat":"m",'Sun':"r"}
sns.catplot(x="day", 
            y="total_bill",
            data=tips, 
            kind="violin",
            split=True,
            hue='sex',
            palette=pal
           )

# Show plot
plt.show()
```

# Input Data: Univariate (1D Series)

## Distribution Plots

### a. distribution plot
<li> Provides a way to look at a univariate distribution. </li>
<li> A univeriate distribution provides a distribution for one variable</li>
<li> Kernal Density Estimation with a Histogram is provided</li>
<li> kde=False removes the KDE</li>
<li> Bins define how many buckets to divide the data up into between intervals</li>
<li> For example put all profits between $10 and $20 in this bucket</li>


```python
sns.distplot(tips['total_bill'], 
             bins=10,
            kde = True
            )
plt.show()
```

```python
sns.kdeplot(tips['total_bill']
            )
plt.show()
```

### b. ecdf Plot

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.ecdfplot(data=penguins, x="flipper_length_mm")
plt.show()
```

### c. Rug Plot

<li> Plots a single column of datapoints in an array as sticks on an axis
<li> With a rug plot you'll see a more dense number of lines where the amount is 
# most common. 
<li>This is like how a histogram is taller where values are more common

```python
sns.rugplot(tips['tip'],palette='pastel', height=0.1)
```

### Styling

```python
# You can set styling for your axes and grids
# white, darkgrid, whitegrid, dark, ticks
sns.set_style('white')

# You can use figure sizing from Matplotlib
plt.figure(figsize=(6,4))

# Change size of lables, lines and other elements to best fit
# how you will present your data (paper, talk, poster)
sns.set_context('paper', font_scale=1.8)

sns.scatterplot(x='total_bill', y='tip', data=tips)

# Get rid of spines
# You can turn of specific spines with right=True, left=True
# bottom=True, top=True
sns.despine(left=False, bottom=False)
```

### Pairplots

<li> To create a heatmap with data you must have data set up as a matrix where variables
are on the columns and rows

<li> Correlation tells you how influential a variable is on the result
<li> Colormaps - https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

```python
cr_df = sns.load_dataset('car_crashes')
cr_df
```

```python
cr_df.corr()
```

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

# So we see that n previous accident is heavily correlated with accidents, while the
# insurance premium is not
crash_mx = cr_df.corr()

# Create the heatmap, add annotations and a color map
sns.heatmap(crash_mx, 
            annot=True, 
            cmap='Blues'
           )
```

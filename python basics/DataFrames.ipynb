### Missing Data

```python
new_dframe = dframe.dropna()
new_dframe
```

```python
#We can also specify to only drop rows that are complete missing all data
dframe.dropna(how='all')
```

```python
#Or we can specify to drop columns with missing data
dframe = DataFrame([[1,2,3,4],[np.nan,5,6,11],[7,np.nan,9,10],[np.nan,np.nan,np.nan,33]])
dframe
```

```python
dframe.dropna(axis=1)
```

```python
#Dropping any rows that dont have at least 2 data points
dframe.dropna(thresh=2)
```

### Merge data

```python
movies = pd.DataFrame({'movieid':[1,2,3,4,5],
                     'title': ['Toy Story2','Mission Impossible','Avatar','Blade Runner','Ghost']
                     })
movies
```

```python
members = pd.DataFrame({'memid':[1,2,3,4],
                     'name': ['Tom','Jon','Aron','Bobby'],
                      'movieid': [1,3,4,np.NaN]
                     })
members
```

```python
pd.merge(movies,members, on='movieid')
```

```python
pd.merge(movies,members, on='movieid', how='left')
```

```python
pd.merge(movies,members, on='movieid', how='right')
```

```python
# Select only 3 columns from the merged dataframe
pd.merge(movies,members, on='movieid', how='right')[['name','title']]
```

```python
pd.merge(movies,members, on='movieid', how='outer')
```

### Handling Duplicates

```python
ddup = DataFrame({'col1':['A','A','B','B','B'], 'col2':[2,2,2,3,3]})
ddup        
```

```python
#Use duplicated to find duplicates
ddup.duplicated()
```

```python
ddup.drop_duplicates()
```

```python
ddup.drop_duplicates(keep='last')
```

```python
## You can filter which duplicates to drop by a single column
## ddup.drop_duplicates(['col1'])
ddup.drop_duplicates(subset=['col1'])
```

## Group By

```python
dgrp = pd.read_csv('Salaries.csv')
dgrp
```

```python
grp = dgrp.groupby('rank')
```

```python
grp
```

```python
for rank,group in grp:
    print(rank,group[['discipline','sex','salary']])
```

```python
grp.get_group('Prof')
```

```python
grp2 = dgrp.groupby(['rank','sex'])
```

```python
for rank, group in grp2:
    print(rank, group)
```

```python
grp2.get_group(('Prof','Female'))
```
